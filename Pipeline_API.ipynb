## Executando a pesquisa

```python
'''
1 - Usando a API pública do mercado livre traga todos os resultados de
uma consulta sobre um termo especifico para o ambiente python.
'''
```

```python
'''
2 - Armazene os resultados dessa busca em um df pandas
'''
```

```python
df = pd.DataFrame() # Criando um df em DataFrame
lista = ['iphone', 'laptop', 'cama', 'terno', 'ferramentas'] # Criando uma lista com 5 produtos as ser pesquisados, limite da API
for b in lista: # Fazendo um for para percorrer a lista e me dar um numero de pesquisa
    busca = b
    a = 0
    b = 20

    for i in range(b):
        print(i)
        # URL da API do Mercado Livre
        url = f"https://api.mercadolibre.com/sites/MLB/search?q={busca}&limit=50&offset={a}"
        # Fazer a requisição GET
        response = requests.get(url)

        # Verificar se a requisição foi bem-sucedida
        if response.status_code == 200:
            # Carregar os dados em formato JSON
            dados = response.json()

            # Extrair os dados relevantes
            produtos = []
            for item in dados.get('results', []):
                produto = {
                    "Título": item.get('title'),
                    "Preço (R$)": item.get('price'),
                    "Link": item.get('permalink'),
                    "Condição": item.get('condition'),
                    "Categoria": item.get('category_id')
                }
                produtos.append(produto)

            # Criar o DataFrame

            df = pd.concat([df, pd.DataFrame(produtos)], ignore_index=True) # concatenando o DataFrame

            # Exibir o DataFrame
            display(df)
            a += 50
        else:
            print(f"Erro na requisição: {response.status_code}")
```

## Fazendo Drop's e alterações nescessarias no DataFrame com Pandas

```python
'''
3 - faça as transformações que julgar necessárias no DataFrame
Ex: mudar nomes de colunas, mudar tipos de dados, remover dados nulos,
criar novas colunas e etc...
'''
```

```python
display(df.head(1)) # Alterando o titulo das colunas
df.rename(columns={'Título': 'titulo',
                   'Preço (R$)': 'preco',
                   'Link': 'link',
                   'Condição': 'condicao',
                   'Categoria': 'categoria'}, inplace=True)
df.head(1)
```

```python
# alterando a descrição dos produtos em condição
df['condicao'] = df['condicao'].replace('new', 'novo')
df['condicao'] = df['condicao'].replace('used', 'usado')
```

```python
#Organizando a coluna
df = df[['titulo', 'preco', 'categoria', 'condicao', 'link']]
df.head()
```

```python
# Ordenando por valor do maior para o menor
df = df.sort_values('preco', ascending=False)
df.head()
```

```python
# Removendo o Index da tabela
display(df.reset_index().drop('index', axis=1).head(1))
```
